```python
#Import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Load the data
Data_Cluster=pd.read_csv("ClusteredDataSet.csv")
```

```python
#Plot the data
plt.scatter(Data_Cluster['Age'],Data_Cluster['Income'],c=Data_Cluster['clusterNo'],cmap='rainbow')
```

```python
#cluster0
cluster0 = Data_Cluster[Data_Cluster["clusterNo"] == 0]
cluster0
```

```python
cluster0["Profession"].unique()
```

```python
count2 = len(cluster0[cluster0["Profession"] == 2])

count6 = len(cluster0[cluster0["Profession"] == 6])

print(count2,count6)

y = np.array([count2,count6])
y
```

# Luxury Vehicles

```python
Professionlabels = ["Engineer", "Executive"]

plt.pie(y, labels = Professionlabels, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.show() 
```

```python
cluster0_Age = cluster0['Age'].mean()
cluster0_Family_Size = cluster0['Family_Size'].mean()
cluster0_Income = cluster0['Income'].mean()

print("Average age ",cluster0_Age)
print("Average Family_Size ",cluster0_Family_Size)
print("Average Income ",cluster0_Income)
```
